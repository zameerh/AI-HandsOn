```python
import matplotlib.pyplot as plt
import numpy as np

# Parameters
initial_investment = 10000
monthly_rate = 0.05
months = np.arange(1, 25)  # 1 to 12 months

# Calculate compound values
values = [initial_investment * (1 + monthly_rate)**n for n in months]

# Create bar chart
plt.figure(figsize=(10, 6))
bars = plt.bar(months, values, color='skyblue', edgecolor='black')

# Add value labels vertically on top of each bar
for bar in bars:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, yval + 500,
             f'£{yval:,.0f}',
             ha='center', va='bottom', rotation=90)

# Customize chart
plt.xlabel('Month')
plt.ylabel('Value (£)')
plt.title('Monthly 5% Compound Growth (Starting at £10,000)')
plt.grid(True, axis='y', linestyle='--', alpha=0.7)
plt.xticks(months)

# Adjust layout to prevent label cutoff
plt.tight_layout()

# Save the plot
plt.savefig('compound_growth_chart.png')
```
